```python
#### TR in localizer is 2 seconds.
#### Therefore there should be a 4 second delay to sycronize the mri and exp program

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from scipy.io import savemat

logging_path = r'C:\Users\Martin\Desktop\Uni\Masterarbeit\Masterarbeit_Datenanalyse\probanden\vp05_ET\logs'

os.chdir(logging_path)
logging = pd.read_csv('ET_localizer.csv', index_col= 0) 
condition_list= list(logging['condition'].value_counts().index)
```

```python
#condition_list.remove('Target')
condition_list
```

```python
logging
```

```python
onset_array = np.empty((3,), dtype= np.object_)
for i in range(len(condition_list)):
    cond = condition_list[i]
    cond_onst = list(logging[logging['condition'] == cond]['time when it onsets'])
    cond_onset_array = np.array([[num] for num in cond_onst])
    cond_onset_array = cond_onset_array + 4
    onset_array[i] = cond_onset_array
    
mat_dict = {}
```

```python
duration_list = [0.20, 0.20, 0.20]

duration_array = np.empty((3,), dtype= np.object_)
for i in range(len(duration_list)):
    duration = duration_list[i]
    duration_array[i] = duration
duration_array    
```

```python
name_list = condition_list
names_array = np.empty((3,), dtype= np.object_)
for i in range(len(name_list)):
    name = name_list[i]
    names_array[i] = name
names_array
```

```python
mat_dict['onsets'] = onset_array
mat_dict['names'] = names_array
mat_dict['durations'] = duration_array
```

```python
savemat('ET_loc_MCF.mat', mat_dict)
```
