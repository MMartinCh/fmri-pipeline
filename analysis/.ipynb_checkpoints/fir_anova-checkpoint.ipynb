```python
from pathlib import Path
import pandas as pd
```

```python
# Get FIR files
base_path = Path.cwd()

brain_region = "left_OFA"
exclude_vp = ["vp02"]

roi_path = base_path / "roi_analysis"/ brain_region
fir_files = [f for f in roi_path.glob(f"*{brain_region}_fir.txt")]

fir_files[:5]
```

```python
# Get questionnaire results
quest_file = base_path / "clean_questionnaire_results.xlsx"
quest_df = pd.read_excel(quest_file)

quest_df.head()
```

```python
# Exclude participants below cutoff
cutoff = 3.5
exclude_vp.extend(quest_df.loc[quest_df["total_avg"] <= cutoff, "idx"].tolist())

exclude_vp
```

```python
# Create main df from FIR files
df = pd.DataFrame(columns= ["idx", "condition", "time_point"])

for subj_i, file in enumerate(fir_files):
    subj_id = file.stem.split("_")[0]
    
    rows = []

    if not subj_id in exclude_vp:
        print(f"Extract time points for {subj_id}")

        with open(file, 'r') as f:
            lines = f.readlines()

            row4 = [float(x) for x in lines[3].split()]
            row5 = [float(x) for x in lines[4].split()]
            row6 = [float(x) for x in lines[5].split()]

            peak = [max(v4, v5, v6) for v4, v5, v6 in zip(row4, row5, row6)]
            rows.append(peak)

        for row in rows:
            for i, v in enumerate(row):
                if i == 0:
                    df.loc[len(df)] = {"idx": subj_id, "condition": "congruent", "time_point": v}
                elif i == 1:
                    df.loc[len(df)] = {"idx": subj_id, "condition": "incongruent", "time_point": v}
                elif i == 2:
                    df.loc[len(df)] = {"idx": subj_id, "condition": "neutral", "time_point": v}

    else:
        print(f"Subject {subj_id} excluded.")

df.head(10)
```

```python
from statsmodels.stats.anova import AnovaRM

# Repeated-measures ANOVA for time_points ~ condition (per subject)
anova = AnovaRM(data = df, 
                depvar= "time_point",
                subject= "idx",
                within= ["condition"]
                )

print(anova.fit())
```

```python
import pingouin as pg

# Post-hoc Pairwise t-Tests
posthoc = pg.pairwise_tests(
    data=df,
    dv="time_point",
    within="condition",
    subject="idx",
    padjust="holm"
)

posthoc
```
